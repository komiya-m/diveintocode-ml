## 【課題1】データの取得
以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。

sklearn.datasets.load_iris — scikit-learn 0.20.2 documentation

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

index = np.arange(0,150) #行のインデックス、別になくても０から順番に入る
iris = load_iris()
X = pd.DataFrame(iris.data,index=index, columns=iris.feature_names) #Dataframe ”X”を作成
#class DataFrame(data=None, index=None, columns=None, dtype=None, copy=False)
```

```python
print(iris) #データセットの情報を見れる
```

```python
y = pd.DataFrame(iris.target, columns=["Species"]) #Dataframe "y"作成
```

```python
X[:5]
```

```python
y[:5]
```

## 【課題2】データの結合
そして、これらX,yを結合させたDataFrame型変数dfを作成してください。これ以降はこのdfを使っていきます。

```python
df = pd.concat([X, y], axis=1) #X,yを横方向に連結 axis=1
```

```python
df[:5]
```

### seabornについて
matplotlibのラッパー  
seabornはnumpy, matplotlib, pandasに依存しているので、同時にimportします。  
参考　https://qiita.com/HiromuMasuda0228/items/babdfa175815fb3e3a94  
本家　https://seaborn.pydata.org/generated/seaborn.boxplot.html

```python
import seaborn as sns

fig, ax = plt.subplots(figsize=(10,7))  #見にくいのでサイズ変更
plt.title("boxplot",{'fontsize':20})
sns.boxplot(data=df,fliersize = 3) #seabornだとラベルが勝手に入る。色が勝手に入る。
# fiersize 外れ値のサイズ
```

### ⑶violinplotを使いデータの分布を可視化する  

```python
fig, ax = plt.subplots(figsize=(10,7))  #見にくいのでサイズ変更
plt.title("violinplot",{'fontsize':20})
sns.violinplot(data=df)
```

### ⑷boxplotとviolinplotは似た図を表示します。
### これらはどう違い、それぞれどのような利点があるのか説明してください。

boxplot図は四分位点、最大値、最小値が明確。  
だが最大値と四分位点、最小値と四分位点の間にもデータはあるが、箱が四分位点で終わってしまっているため、  
あたかもそこにはデータが無いような印象を受ける。  
violinplot図では中央値、四分位点のほかに、データの分布密度も同時に確認でき、情報量がboxplotよりも多い。だが知名度が低い。

### ⑶４つの特徴量に関する相関係数行列を作る  

相関係数について  
参考　https://atarimae.biz/archives/7966  
相関係数は一般的に、＋1 に近ければ近いほど「強い正の相関がある」、  
−1 に近ければ近いほど「強い負の相関がある」、0 に近ければ近いほど「ほとんど相関がない」と評価されます。  

注意点、①少ないデータからの相関係数はあまり意味をなさない  
②外れ値に弱い   ③相関関係があるからといって因果関係があるとは限らない

参考　https://note.nkmk.me/python-pandas-corr/  
相関係数の算出方法の指定: 引数method  
'pearson': ピアソンの積率相関係数（デフォルト）  
'kendall': ケンドールの順位相関係数  
'spearman': スピアマンの順位相関係数

```python
#pandasで作成
df_corr = df.iloc[:,:4].corr(method='pearson')
```

```python
df_corr
```

### ⑷相関係数行列のヒートマップを作る

```python
sns.palplot(sns.color_palette("cubehelix", 24))
fig, ax = plt.subplots(figsize=(8,8))
sns.heatmap(df_corr, square=True, annot=True, cmap="cubehelix") # annot=True:格子の中に値を表示
```
